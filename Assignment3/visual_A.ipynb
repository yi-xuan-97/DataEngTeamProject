```python
import pandas as pd
import os
import folium
from datetime import date
import random

today = str(date.today())
filepath = '/Users/zhengmao/Documents/Github/DataEngTeamProject/Assignment3/2022-05-21.csv'
df = pd.read_csv(filepath)
df = df[['y_coordinate', 'x_coordinate']]	
basicMap = folium.Map(location=[df['y_coordinate'].mean(), df['x_coordinate'].mean()], zoom_start = 11, tiles='cartodbpositronnolabels')
folium.PolyLine(locations=[[45.592404, -122.550711],[45.586158, -122.541270]]).add_to(basicMap)
filespath = '/Users/zhengmao/Documents/Github/DataEngTeamProject/Assignment3/validData/2022-05-21/'
filelist = os.listdir(filespath)
for file in filelist:
    df = pd.read_csv(filespath + file)
    if len(df) != 0:
        color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])]
        df.sort_values(by='arrive_time', inplace=True)
        route_number = set(df['route_number'])
        folium.PolyLine(locations=df[['y_coordinate', 'x_coordinate']], color = color, popup=route_number).add_to(basicMap)

basicMap.fit_bounds(bounds=[[45.592404, -122.550711],[45.586158, -122.541270]], max_zoom=13)
basicMap.save(f'Pages/{today}-A.html')
basicMap
```
