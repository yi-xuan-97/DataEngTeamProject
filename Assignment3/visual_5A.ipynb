```python
import pandas as pd
import os
import folium
import folium.plugins
from datetime import date

df = pd.read_csv('2022-05-21.csv')
# df = df[['y_coordinate', 'x_coordinate']]	
df2 = pd.read_csv('2022-05-22.csv')
basicMap = folium.plugins.DualMap(location=[df['y_coordinate'].mean(), df['x_coordinate'].mean()], zoom_start = 11, tiles='cartodbpositronnolabels')
def addDFtoMapi(df, mapi):
    df[['hour', 'min', 'sec']] = df.arrive.str.split(':', expand=True)
    df['count'] = 1
    # 'y_coordinate', 'x_coordinate'
    df_hour_list = [] # create blank list
    # 2020-10-25
    # loop through each hour
    for hour in df['hour'].sort_values().unique(): 
        df_hour_list.append(df.loc[df['hour'] == hour, # for each hour append to list  
        ['y_coordinate', 'x_coordinate', 'count']]\
        .groupby(['y_coordinate', 'x_coordinate'])\
        .sum().reset_index().values.tolist()) #sum totals per station, reset index and create list
    folium.plugins.HeatMapWithTime(df_hour_list,radius=5,auto_play=True, position='bottomright').add_to(mapi)
addDFtoMapi(df, basicMap.m1)
addDFtoMapi(df2, basicMap.m2)
basicMap.save('Pages/ForProblem5A.html')
basicMap
```
