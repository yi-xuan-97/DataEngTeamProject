```python
import pandas as pd
import os
import folium
from datetime import date

today = str(date.today())

path = '/Users/zhengmao/Documents/Github/DataEngTeamProject/Assignment3/validData/'
filelist = os.listdir(path)
workTime = {}
for file in filelist:
    df = pd.read_csv(path + file)
    if len(df) != 0:
        route_number = list(set(df['route_number']))[0]
        maxari = df['arrive_time'].max()
        minari = df['arrive_time'].min()
        # print(route_number)
        workTime[str(route_number)] = maxari-minari
        # workTime.append(maxari-minari)
print(workTime)
maxRoute = max(workTime, key=workTime.get)
print(maxRoute)
df = pd.read_csv('2022-05-22.csv')
df.drop(df[df['route_number']!= int(maxRoute)].index, inplace = True)
df.sort_values(by='arrive_time', inplace=True)
basicMap = folium.Map(location=[df['y_coordinate'].mean(), df['x_coordinate'].mean()], zoom_start = 12, tiles='cartodbpositronnolabels')
folium.PolyLine(locations=df[['y_coordinate', 'x_coordinate']]).add_to(basicMap)
basicMap.save('Pages/forProblemD.html')
basicMap

```
