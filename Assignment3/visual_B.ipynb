```python
import pandas as pd
import os
import folium
import folium.plugins
from datetime import date

today = date.today()
df = pd.read_csv('2022-05-22.csv')
df = df.loc[df['route_number'] == 65]
# df = df.drop(df[(df.leave_time	 < 57600) & (df.leave_time	 > 64800)].index)
df.drop(df[df['leave_time'] < 57600].index, inplace = True)
df.drop(df[df['leave_time'] > 64800].index, inplace = True)
df.sort_values(by='leave_time', inplace=True)
basicMap = folium.Map(location=[df['y_coordinate'].mean(), df['x_coordinate'].mean()], zoom_start = 12, tiles='cartodbpositronnolabels')
df1 = df.drop(df[df['direction'] == 1].index)
df0 = df.drop(df[df['direction'] == 0].index)
df2 = df[['y_coordinate', 'x_coordinate']]
# df1 = df2.drop(df[df['leave_time'] < 57600].index)

# # 4PM = 57600	# 6PM = 64800
folium.PolyLine(locations=[[45.592404, -122.550711],[45.586158, -122.541270]], color='black').add_to(basicMap)

folium.PolyLine(locations=df1[['y_coordinate', 'x_coordinate']], color= 'red').add_to(basicMap)
folium.PolyLine(locations=df0[['y_coordinate', 'x_coordinate']], color= 'blue').add_to(basicMap)
# folium.plugins.PolyLineOffset(df1[['y_coordinate', 'x_coordinate']], color="#080", opacity=1, offset=10).add_to(basicMap)
basicMap.save(f'Pages/{today}-B.html')
basicMap
# df
# set(df['vehicle_number'])
```
